### Aula 10 - Conversão de Tipos de Dados - Parte 01



*  int() - converte para inteiro

```python
nova_altura = int(altura)
print(nova_altura)
print(type(nova_altura))
print(type(altura))
```

### Aula 12 Operadores Aritméticos

```python
# Operação de soma
num1 = 10
num2 = 30
soma = num1 + num2
print("A soma é :", soma)
```

```python
# Subtração
diferanca = num1 - num2
print ("A Diferença é:", diferanca)
```

```python
# Multiplicação
produto = num1 * num2
print("O produto é:", produto)
```

```python
#  Divisão real ** sempre vai ser um Número Real
divisao_real = num1 / num2
print("A divisão é:", divisao_real)
# Tipo da variavel resultante de uma divisão Real
print("O tipo da divisão real é: ", type(divisao_real))
num3 = 2
divisao_real2 = num1 / num3
print("A divisão é:", divisao_real2)
```
